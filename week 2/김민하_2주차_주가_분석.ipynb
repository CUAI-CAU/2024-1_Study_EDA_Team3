#### 일별 수익률 구하기
    일별 수익률(%): 하루 동안의 수익률

    = (오늘 가격 - 어제 가격) / 어제 가격 * 100

    - 양수 수익률: 자산 상승
    - 음수 수익률: 자산 하락
    - 0 수익률: 자산의 가격 변하지 않았음

```python
df = fdr.DataReader(samsung_code, '2015-01-01')
df.head()
```

```python
# 종가 기준 일별 수익률 계산
# pct_change(periods=1, fill_method='pad', limit=None, freq=None, kwargs)
# 한 객체 내에서 행과 행의 차이를 현재값과의 백분율로 출력하는 메서드 > (다음행-현재행) / 현재행
df = (df['Close'].pct_change()) * 100
```

#### 주가 등락 구하기
    # 이항분포 적용
    Q. 주가 데이터를 활용하여 5일 중 3일 동안 주가가 오를 확률?

    A. 주가가 오를 확률 p=0.42 가정
       5일(n=5) 중 3일(k=3)동안 주가가 오를 확률
       => P(X=3) = 5!/(3!2!) * (0.42)^3 * (1-0.42)^2

```python
df = fdr.DataReader('005610', '2010-01-01')
df.head()
```

```python
df1 = (df.loc['2023-01-01':'2023-12-31']).copy()
df1
```

```python
df1['ret'] = (df1['Close'].pct_change())*100
df2 = df1['ret'] # SPC삼림 2023 일별수익률
r = df2.dropna() # 결측치 행 제거
```

```python
len(r)
```

```python
positive = 0 # 1년 중 양수 수익률이었던 날의 수
n = len(r)
for i in range(n):
    if r[i] > 0: positive +=1
negative = n - positive
```

```python
positive
```

```python
positive_rate = positive / n
positive_rate # 양수 수익률 비율 > 주가가 오를 확률
```

```python
# 5일 중 3일 간 주가가 오를 확률
from math import factorial
c = factorial(5) / (factorial(3) * factorial(2))
p = (positive_rate**3) * (1-positive_rate)**2
prob = c * p
prob
```
