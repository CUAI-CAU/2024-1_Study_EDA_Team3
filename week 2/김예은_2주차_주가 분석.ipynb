### 5. 일별 수익률 계산 및 시각화

```python
df = fdr.DataReader(sm_code, '2015-01-01')
df.head()
```

```python
# 주식 데이터프레임(df)에서 'Close' 열의 데이터를 추출하고, 이를 사용하여 일별 수익률을 계산하는 코드입니다.
# .pct_change(): 일별 수익률을 계산하는 과정으로, 현재 값과 이전 값 사이의 백분율 변화를 나타냅니다
change_df = (df['Close'].pct_change()) * 100
change_df
```

```python
change_df.describe()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(change_df.index, change_df)
plt.title('Daily Returns of SM Stock')
plt.xlabel('Date')
plt.ylabel('Daily Returns (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(change_df, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Returns (%)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 6. 증가등락 구하기

```python
df = fdr.DataReader(sm_code,'2015-01-01')
df.head(5)
```

```python
df1 = df.loc['2020-01-01':'2023-12-31']
df1['ret'] = (df1['Close'].pct_change())*100
df2 = df1['ret']
r = df2.dropna()
len(r)
```

```python
positive = 0  # 986일 중 일별 수익률이 양수인 날의 개수
n = len(r)

for i in range(n):
  if r[i] > 0:
    positive += 1

negative = n - positive  # 986일 중 일별 수익률이 음수인 날의 개수
```

```python
# 일별 수익률이 양수일 확률
positive_rate = positive / n
positive_rate
```

```python
import math

# 3일 중에 2일 주가가 오를 확률
a = math.factorial(3)
b = math.factorial(2) * math.factorial(1)
c = ((positive_rate)**2) * ((1 - positive_rate)**1)

prob = a / b * c
prob
```
